Merging, Concatenating & Joining Data

```python
df1 = pd.DataFrame({
    'Item': ['A', 'B', 'C'],
    'Price': [10, 20, 30]
})

df2 = pd.DataFrame({
    'Item': ['D', 'E', 'F'],
    'Price': [40, 50, 60]
})

```

```python
pd.concat([df1, df2]).reset_index().drop('index', axis = 1)
```

```python
df1 = pd.DataFrame({
    'Item': ['A', 'B', 'C'],
    'Price': [10, 20, 30]
})

df3 = pd.DataFrame({
    'Country': ['USA', 'Canada', 'Mexico'],
    'Availabe': [True, True, False]
})
```

```python
pd.concat([df1, df3], axis = 1)
```

```python
df4 = pd.DataFrame({
    'Item': ['A', 'B', 'C'],
    'Country': ['USA', 'Canada', 'Mexico']
})
df4
```

```python
df1
```

```python
pd.merge(df1, df4, how = 'inner')
```

```python
df5 = pd.DataFrame({
    'Price': [10, 20, 30],
}, index = ['A', 'B', 'C'])
df6 = pd.DataFrame({
    'Country': ['USA', 'Canada', 'Mexico'],
}, index = ['B', 'C', 'D'])
```

```python
df4.join(df5, how = 'inner')
```
